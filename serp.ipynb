```python
def keywords(url):
    adv.crawl(
                url_list=[url],
                output_file="output/crawl_output.jl",
                )
    crawlDf = pd.read_json("output/crawl_output.jl",lines=True)
    body_text = crawlDf["body_text"][0]
    print(body_text)
    body_text = body_text.replace(list(stopwords["english"]),"")
    body_text = body_text.strip()
    print(body_text)

keywords("https://gerwholesalers.com/")
```

```python
def analyzeCrawlLogs():
    logsDf = adv.crawllogs_to_df(logs_file_path="output_file.log")

    logs_m = logsDf["message"].value_counts().to_json()
    logs_s = logsDf["status"].value_counts().to_json()
    logs_mi = logsDf["middleware"].value_counts().to_json()

    logsDf = logsDf.reset_index(drop=True).to_html(
        classes="table", justify="center"
    )

    logsDf = logsDf.replace(
        'class="dataframe table"',
        'class="table table-primary table-striped text-center"',
    )
    
    
    return {
        'logs_message': logs_m,
        'logs_status': logs_s,
        'logs_mi': logs_mi,
        'logs_dt': logsDf 
    }
```
