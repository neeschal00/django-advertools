```python
import networkx as nx
from urllib.parse import urlparse
```

```python
url ='https://www.example.com/products?id=123&category=electronics#details'
# Parse URL
parts = urlparse(url)
# Create graph
G = nx.DiGraph()
# Add path parts as nodes
nodes = parts.path.split('/')
G.add_nodes_from(nodes)
# Add query parameters as child nodes
for param in parts.query.split('&'):
    key, value = param.split('=')
    G.add_edge(parts.path, key)
    G.add_edge(key, value)
    # Add fragment as child node
if parts.fragment:
    G.add_edge(parts.path, parts.fragment)
# Draw tree
pos = nx.nx_agraph.graphviz_layout(G, prog='neato')
nx.draw(G, pos, with_labels=True)
```
