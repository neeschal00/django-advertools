## Get Keywords Funtion
The function below that uses regular expression to sort and clean any **numeric values** and **symbols**, also removing the stopwords consisted in the text to get keywords existing in the function which will be used in the application for analytics purposes.

```python
def extract_keywords(body_text):    
    pattern = r'[^a-zA-Z0-9@\s]'
    body_text = re.sub(pattern,"",body_text)
    numPattern = r'\d+'
    body_text = re.sub(numPattern,"",body_text)
    body_text = re.sub(r'\b\w\b', '', body_text)
    whiteSpacePattern = r'\s+'
    body_text = re.sub(whiteSpacePattern," ",body_text)
    for text in adv.stopwords["english"]:
        body_text = body_text.replace(" "+text.lower()+" "," ")
    keywords = body_text.lower().split(" ")

    return keywords
```

### Text Readability Score of the content in the webpage

Get the text readability score of the content of the webpage using **Flesch-Kincaid Readability Formula** which commonly used. The nltk libraries is used which must be downloaded and set

```python
def text_readability(text):
    sentences = nltk.sent_tokenize(text)
    words = nltk.word_tokenize(text)
    
    num_sentences = len(sentences)
    num_words = len(words)
    num_syllables = 0
    
    for word in words:
        num_syllables += syllable_count(word)
    
    if num_sentences == 0:
        num_sentences = 1
    
    readability_score = 206.835 - 1.015 * (num_words / num_sentences) - 84.6 * (num_syllables / num_words)
    return readability_score

def syllable_count(word):
    vowels = "AEIOUaeiou"
    syllables = 0
    prev_char = None
    
    for char in word:
        if char in vowels and (prev_char is None or prev_char not in vowels):
            syllables += 1
        prev_char = char
    
    if word.endswith('e'):
        syllables -= 1
    
    if syllables == 0:
        syllables = 1
    
    return syllables
```
